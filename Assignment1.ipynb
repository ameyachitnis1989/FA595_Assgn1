```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#define plot parameters and functions:
plt.figure(figsize=(10,5))
x=np.linspace(1,10,1000)
y1=np.sin(x)
y2=np.cos(x)

#plot sine and cosine
plt.plot(x,y1,label = 'sine')
plt.plot(x,y2,label = 'cosine')

#label axes
plt.xlabel("X")
plt.xlabel("Y")

#display plot
plt.legend()
plt.show()
```
